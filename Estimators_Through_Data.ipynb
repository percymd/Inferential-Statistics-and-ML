## Distribuciones muestrales

```python
import matplotlib.pyplot as plt
from IPython.core.display import Image
import seaborn as sns
%matplotlib inline
```

### T Student

```python
from scipy.stats import t

data1 = t.rvs(100, size=10000000)
data2 = t.rvs(40, size=10000000)

ax = sns.histplot(data1, bins=500, kde=False, color='blue')
ax = sns.histplot(data2, bins=500, kde=False, color='red')
```

### Chi Cuadrada

```python
from scipy.stats import chi2

data1 = chi2.rvs(5, size=10000000)
data2 = chi2.rvs(15, size=10000000)

ax = sns.histplot(data1, bins=500, kde=False, color='blue')
ax = sns.histplot(data2, bins=500, kde=False, color='red')
```

### Fisher

```python
from scipy.stats import f

data1 = f.rvs(5, 25, size=10000000)
data2 = f.rvs(15, 25, size=10000000)

ax = sns.histplot(data1, bins=500, kde=False, color='blue')
ax = sns.histplot(data2, bins=500, kde=False, color='red')
```

### Función de densidad de probabilidad (Probability density function)


```python
f.pdf(4, 15, 25)
```

### Función de distribución (Cumulative distribution function)

```python
f.cdf(4, 15, 25)
```

### Percent Point Function (Inverse of the CDF)

```python
f.ppf(0.9988900520089906, 15, 25)
```

```python
f.ppf(0.5, 15, 25)
```
