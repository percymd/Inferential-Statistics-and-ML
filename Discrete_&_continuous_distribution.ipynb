## Funciones de distribuci√≥n discreta y continua

```python
import scipy.stats
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
from scipy.stats import binom


p = 0.3
n = 8 

x = list(range(0,9,1))
y = list([])

for i in x:
    y.append(binom.pmf(i, p=p, n=n))

fig, ax = plt.subplots()
ax.bar(x,y)
ax.set_ylabel('Probabilidad discreta')

np.sum(y)
```

```python
media = np.average(x, weights=y)
varianza = np.average(((x-media)**2), weights=y)

media, varianza
```

```python
fig, ax = plt.subplots()
ax.bar(x,y)
ax.set_ylabel('Probabilidad discreta')
ax.axvline(x=media, c='red', linestyle='--', label='Valor Esperado')
ax.axvline(x=media + 3*np.sqrt(varianza), c='red', linestyle='--', label='Desviaci√≥n Estandar')
ax.legend()
```

```python
N = 10000

x = list(range(0,N+1,1,))
y = list([])

for i in x:
    y.append(binom.cdf(i, p=0.3, n=N))
    
fig, ax = plt.subplots()
ax.plot(x, y)
ax.set_title('Probabilidad Continua')
```

## Distribuciones de mayor aplicaci√≥n discretas

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from scipy.stats import bernoulli
p = 0.3
data = bernoulli.rvs(p, size=100)
len(data)
```

```python
mean, var, skev, kurt = bernoulli.stats(p, moments='mvsk')
mean, var, skev, kurt
```

```python
import seaborn as sns

#original: sns.distplot
#But it is replace for displot or histplot
ax = sns.histplot(data, bins=30, kde=False, color='blue')
ax.set(xlabel = 'Distribucion Bernoulli', ylabel = 'Frecuencia')
ax.axvline(x=mean, linestyle='--', label='Media')
ax.legend()
```

```python
from scipy.stats import binom
p = 0.3
n = 10
data = binom.rvs(p=p,n=n, size=100)

mean, var, skev, kurt = binom.stats(p=p, n=n, moments='mvsk')
mean, var, skev, kurt
```

```python
import seaborn as sns

ax = sns.histplot(data, bins=30, kde=False, color='#537ac2')
ax.set(xlabel = 'Distribucion Binomial', ylabel = 'Frecuencia')
ax.axvline(x=mean, linestyle='--', label='Media', color='blue')
ax.legend()
```

```python
from scipy.stats import nbinom
p = 0.3
n = 10
data = nbinom.rvs(p=p,n=n, size=100)

mean, var, skev, kurt = binom.stats(p=p, n=n, moments='mvsk')
mean, var, skev, kurt
```

```python
import seaborn as sns

ax = sns.histplot(data, bins=30, kde=False, color='#537ac2')
ax.set(xlabel = 'Distribucion Binomial Negativa', ylabel = 'Frecuencia')
ax.axvline(x=mean, linestyle='--', label='Media', color='blue')
ax.legend()
```

```python
from scipy.stats import poisson
lambda_p = 100
data = poisson.rvs(mu=lambda_p, size=100)

mean, var, skev, kurt = poisson.stats(mu=lambda_p, moments='mvsk')
mean, var, skev, kurt
```

```python
import seaborn as sns

ax = sns.histplot(data, bins=30, kde=False, color='#537ac2')
ax.set(xlabel = 'Distribucion Poisson', ylabel = 'Frecuencia')
ax.axvline(x=mean, linestyle='--', label='Media', color='blue')
ax.legend()
```
